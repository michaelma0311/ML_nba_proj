```python
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=pd.errors.PerformanceWarning)
```

```python
df = pd.read_csv("nbabs.csv", index_col=0)
```

```python
df = df.sort_values("date")

df = df.reset_index(drop=True)
```

```python
del df["mp.1"]
del df["mp_opp.1"]
del df["index_opp"]
```

```python
df
print("team" in df.columns)
```

```python
def add_target(group):
    group["target"] = group["won"].shift(-1)
    return group
df["team1"] = df["team"]
df = df.groupby("team1", group_keys = False).apply(add_target, include_groups=False).reset_index(drop=False)
```

```python
df[df["team"] == 'GSW']
```

```python

df.loc[pd.isnull(df["target"]), "target"] = 2
```

```python
df["target"] = df["target"].astype(int, errors="ignore")
nulls = pd.isnull(df)
nulls = nulls.sum()
nulls = nulls[nulls > 0]
valid_cols = df.columns[~(df.columns.isin(nulls.index))]
```

```python
df = df[valid_cols].copy()


```

```python
from sklearn.linear_model import RidgeClassifier
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import TimeSeriesSplit

rr = RidgeClassifier(alpha=1)

split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, 
                                n_features_to_select=5, 
                                direction="forward",
                                cv=split,
                               )
```

```python
removed_columns = ["season", "date", "won", "target", "team", "team_opp"]
```

```python
selected_columns = df.columns[~df.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
df
```

```python
sfs.fit(df[selected_columns], df["target"])
```

```python
predictors = list(selected_columns[sfs.get_support()])
for i in predictors:
    print(i)
```

```python
def backtest(data, model, predictors, start=2, step=1):
    all_predic = []
    seasons = sorted(data["season"].unique())
    for i in range(start, len(seasons), step):
        cursea = seasons[i]
        train = data[data["season"] < cursea]
        test=data[data["season"] == cursea]
        model.fit(train[predictors], train["target"])
        preds = pd.Series(model.predict(test[predictors]), index=test.index)
        comb = pd.concat([test["target"], preds], axis=1)
        comb.columns = ["actual", "prediction"]
        all_predic.append(comb)
    return pd.concat(all_predic)
```

```python
predictions = backtest(df, rr, predictors)
```

```python
print(predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
predictions = predictions[predictions["actual"] != 2]
accuracy_score(predictions["actual"], predictions["prediction"])
```

```python
df_rolling = df[list(selected_columns) + ["won", "team", "season"]]
```

```python
def find_team_averages(cur):
    rolling = cur.rolling(15).mean()
    return rolling
```

```python
df_rolling = df_rolling.groupby(["team", "season"], group_keys=False).apply(find_team_averages, include_groups = False)
```

```python
df_rolling
```

```python
rolling_cols = [f"{col}_10" for col in df_rolling.columns]
df_rolling.columns = rolling_cols

df = pd.concat([df, df_rolling], axis=1)
```

```python
df
```

```python
df = df.dropna()
df
```

```python
def shifts(team, col_name):
    next_col = team[col_name].shift(-1)
    return next_col
def add_col(df, col_name):
    return df.groupby("team", group_keys=False).apply(lambda x: shifts(x, col_name), include_groups=False)
#print(add_col(df, "home"))
df.loc[:, 'home_next'] = add_col(df, 'home')
df.loc[:, 'team_opp_next'] = add_col(df, 'team_opp')
df.loc[:, 'date_next'] = add_col(df, 'date')
```

```python
full = df.merge(df[rolling_cols + ["team_opp_next", "date_next", "team"]], left_on = ["team", "date_next"], right_on=["team_opp_next", "date_next"])
full
```
